```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
## Dataframe containing data from the inflation_rates csv

df = pd.read_csv('Inflation_rates.csv')

```

```python
# function to get the data from the df data frame
def get_data(years):
    
    for id, year in enumerate(years):
        
        mask = df['YEAR']==year
        new_df = df[mask]

        if id==0:        
            data_df = new_df
        else: 
            data_df = pd.concat([data_df, new_df])
            
    return data_df
        
```

```python
## Fetching the data from the inflation rate dfs

gulf_df = get_data([1990, 1991])

uk_rus_df = get_data([2021, 2022])

iraq_df = get_data([2003, 2004, 2005, 2006,2007, 2008,2009,2010, 2011])
```

```python
## Setting 'YEAR' as the index 
gulf_df.set_index('YEAR', inplace = True)
uk_rus_df.set_index('YEAR', inplace = True)
iraq_df.set_index('YEAR', inplace= True)

## transposing the datasets for plotting
gulf_df = gulf_df.transpose()
uk_rus_df = uk_rus_df.transpose()
iraq_df = iraq_df.transpose()

## Dropping the average value on the dataset. 

uk_rus_df=uk_rus_df.drop('AVE')
gulf_df=gulf_df.drop('AVE')
iraq_df = iraq_df.drop('AVE')
```

```python
## Plotting the dataframes them.

print("Line plots showing the ")
gulf_df.plot()

uk_rus_df.plot(yticks = range(0, 12))

plt.show()
```

```python
#### visualizing the iraq war inflation rates. 
print("Line Chart showing Inflation rates during the Iraq war periods")
iraq_df.plot()
plt.show()
```

```python
### Interest rates 
interest_df = pd.read_csv('Interest_rates.csv')

## getting the interest rates during the gulf war
gulf_war = interest_df[['1990', '1991']]

print("Plot showing the Interest rates during the Iraq war")

## transposing the the dataframe for plotting
gulf_war_df = gulf_war.transpose()

## renaming the 0 columns 
gulf_war_df.rename(columns = {0: 'Interest rates'}, inplace =True)

gulf_war_df.plot(kind='bar')

plt.show()
```

```python
## getting the interest rates during the Iraq war
iraq_war = interest_df[['2003', '2004', '2005', '2006','2007', '2008','2009','2010', '2011']]

print("Plot showing the Interest rates during the Iraq war")

## transposing the dataframe for plotting
iraq_war_df = iraq_war.transpose()
iraq_war_df.rename(columns = {0: 'Interest rates'}, inplace =True)

iraq_war_df.plot()
plt.show()
```

```python
##  Real Interest rates for 2022 are not yet out for world bank 
## Real Interest rates are provided on an yearly granularity. 
## https://data.worldbank.org/indicator/FR.INR.RINR?end=2021&locations=US&start=2021&view=map

uk_russia_df = df[['2021']]

uk_russia_df = uk_russia_df.transpose()

uk_russia_df.rename(columns = {0: 'Interest rates'}, inplace =True)
print("Plot showing interest rates for 2021")
uk_russia_df.plot(kind = 'bar')
plt.show()
```
